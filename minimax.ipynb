```python
search_depth = 3

class AlphaBetaMinimax:
    def __init__(self, state, player):
        self.root = state
        self.player = player
    
    def strategy(self, state):
            static_evaluation = 100 * (np.sum(state == 2) - np.sum(state == 1)) / (np.sum(state==1) + np.sum(state==2))        
            return static_evaluation

    def minimax(self, depth, state, alpha, beta, piece):
        initial_state = state
        valid_moves = __valid_positions__(state, piece)
        if depth == 0 or not valid_moves:
            return self.strategy(state)
                
        else:
            children = [place_piece(state, move[0], move[1], piece) for move in valid_moves]
            if piece == 2:
                maxEval = -np.inf
                for child in children:
                    eval = self.minimax(depth - 1, child, alpha, beta, switch_piece(piece))
                    maxEval = max(maxEval, eval)
                    alpha = max(alpha, eval)
                    if beta <= alpha:
                        break
                return maxEval
            elif piece == 1:
                minEval = np.inf
                for child in children:
                    eval = self.minimax(depth - 1, child, alpha, beta, switch_piece(piece))
                    minEval = min(minEval, eval)
                    beta = min(beta, eval)
                    if beta <= alpha:
                        break
                return minEval
    
    def evaluate_position(self, state):
        state_ini = state
        valid_moves = __valid_positions__(state, self.player)
        best_score = -np.inf
        best_move = None   
            
        for move in valid_moves:
            state = place_piece(state, move[0], move[1], self.player)
            score = self.minimax(search_depth, state, -np.inf, np.inf, switch_piece(self.player))
            if score > best_score:
                best_score = score
                best_move = move
            state = state_ini
        return best_move
    
    def return_move(self):
        move = self.evaluate_position(self.root)
        print("Move played : ", (move[0], move[1]))
        return move
```
